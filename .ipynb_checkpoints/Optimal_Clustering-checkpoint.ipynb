```julia
function read_data(file, scaling)
    # normalized_root = "./normalized_data"
    # scaled_root = "./scaled_data"
    
    if scaling == "normalized"
        data = CSV.read("./normalized_data/$file", DataFrame)
        for c in names(data, Any)
            data[!, c]= Float64.(data[!, c])
        end
        return data
    else
        data = CSV.read("./scaled_data/$file", DataFrame)
        for c in names(data, Any)
            data[!, c]= Float64.(data[!, c])
        end
        return data
    end
end

function k_means_cluster(data, k, seed=0)
    # Documentation: https://clusteringjl.readthedocs.io/en/latest/kmeans.html
    
    data_matrix = Matrix(data)
    kmeans_result = Clustering.kmeans(data_matrix', k; rng=StableRNG(seed))
        # kmeans_result.totalcost (within-cluster sum of square distance)
        # a = assignments(kmeans_result) # get the assignments of points to clusters
        # c = counts(kmeans_result) # get the cluster sizes
        # M = kmeans_result.centers # get the cluster centers
    return kmeans_result
end
```

### Different Measures of Distance:

```julia
function l1_norm(data)
    n = nrow(data)
    distance_matrix = zeros(n, n)

    for i in 1:n
        for j in i:n  #only calculate upper-triangle of values
            obs_i = Vector(data[i, :])
            obs_j = Vector(data[j, :])
            distance_matrix[i, j] = Float64.(sum( abs.( obs_i .- obs_j) ))
        end
    end
    
    return distance_matrix
end

function l2_norm(data)
    n = nrow(data)
    distance_matrix = zeros(n, n)

    for i in 1:n
        for j in i:n  #only calculate upper-triangle of values
            obs_i = Vector(data[i, :])
            obs_j = Vector(data[j, :])
            distance_matrix[i, j] = Float64.(norm( obs_i .- obs_j ))
        end
    end
    
    return distance_matrix
end

function infinity_norm(data)
    n = nrow(data)
    distance_matrix = zeros(n, n)

    for i in 1:n
        for j in 1:n  #only calculate upper-triangle of values
            obs_i = Vector(data[i, :])
            obs_j = Vector(data[j, :])
            distance_matrix[i, j] = Float64.(maximum( abs.( obs_i .- obs_j ) ))
        end
    end
    
    return distance_matrix
end

function get_distances(data, distance)
    if (distance == "l1") | (distance == "manhattan")
        return l1_norm(data)
    elseif (distance == "l2") | (distance == "euclidean")
        return l2_norm(data)
    elseif (distance == "inf")
        return infinity_norm(data)
    end
end
        
```

```julia
function opt_kmeans_v1(data, k_max)
    # Setup necessary ranges and intervals
    n = size(data)[1]
    p = size(data)[2]
    observation_range = 1:n
    variable_range = 1:p
    cluster_range = 1:k_max

    ## Init Model
    model = Model(optimizer_with_attributes(Gurobi.Optimizer, "TimeLimit" => 30.00))
    
    ## Decision Variables
    @variable(model,
        z[i in observation_range, k in cluster_range], Bin)
    @variable(model,
        y[i in observation_range, j in observation_range, k in cluster_range], Bin)

    ## Constriants
    @constraint(model,
        mandatory_assignment[i in observation_range],
        sum( z[i, k] for k in cluster_range) == 1)
    
    @constraint(model,
        within_cluster_1[i in observation_range, j in (i+1):n, k in cluster_range],
        y[i, j, k] <= z[i, k])
    @constraint(model,
        within_cluster_2[i in observation_range, j in (i+1):n, k in cluster_range],
        y[i, j, k] <= z[j, k])
    @constraint(model,
        within_cluster_3[i in observation_range, j in (i+1):n, k in cluster_range],
        y[i, j, k] >= z[i, k] + z[j, k] - 1)
    @constraint(model,
        within_cluster_4[i in observation_range, j in 1:i, k in cluster_range],
        y[i, j, k] == 0)

    ## Objective Function
    @objective(model,
        Min,
        sum( sum( sum( (data[i, j]*y[i, j, k]) for j in (i+1):n) for i in observation_range) for k in cluster_range))

    ## Solve
    optimize!(model)

    return model
end
```

```julia
function evaluate_clustering(data, assignments, centers)
    data_matrix = Matrix(data)
    K = maximum(assignments)
    n = size(data_matrix)[1]
    p = size(data_matrix)[2]

    wcssd = zeros(K)

    for k in 1:K
        cluster_obs = data_matrix[assignments .== k, :]
        for i in 1:size(cluster_obs)[1]
            wcssd[k] = wcssd[k] + norm(cluster_obs[i, :] .- centers[k, :])
        end
    end

    return sum(wcssd)
end
```

### Abalone Dataset:

#### Abalone (Normalized):

```julia
abalone_normalized = read_data("abalone_z.csv", "normalized");

# Pre-calculate distance measures between observations
@time begin
    abalone_norm_d = get_distances(abalone_normalized, "l2");  # n x n matrix of positive distance values
end

# L1 took ~37 seconds
# L2 took ~36 seconds
# L-inf took ~75 seconds
```

```julia
data_matrix = Matrix(abalone_normalized)

measure = []
K_max = 20
for k in 1:K_max
    Random.seed!(k)
    kmeans_result = Clustering.kmeans(data_matrix', k; rng=StableRNG(k))
    push!(measure, sum(kmeans_result.totalcost))
end

plot(1:K_max, measure, xlabel="# of Clusters", ylabel="Within Cluster Sum of Squares", title="Elbow Plot for KMeans Clustering", linewidth=2)
```

```julia
@time begin
    abalone_z_kmeans = k_means_cluster(abalone_normalized, 3);
end

# Heuristic KMeans took ~3 seconds
println("Clustering objective value: $(sum(abalone_z_kmeans.total_cost))")
evaluate_clustering(abalone_normalized, assignments(abalone_z_kmeans))
```

```julia
similarity_normalized = read_data("similarity_prediction_z.csv", "normalized");
```

```julia
distance = "l1"  # "l1", "l2", or "inf"
data = abalone_normalized
k_max = 3

n = size(data)[1]
p = size(data)[2]

observation_range = 1:n
variable_range = 1:p
cluster_range = 1:k_max

model = Model(optimizer_with_attributes(Gurobi.Optimizer, "TimeLimit" => 60.00))

@variable(model,
    c[k in cluster_range, j in variable_range])
@variable(model,
    z[i in observation_range, k in cluster_range], Bin)

@constraint(model,
    ensure_assignment[i in observation_range],
    sum( (z[i, k]) for k in cluster_range) == 1)

@constraint(model,
    cluster_numbers[k in cluster_range],
    sum( (z[i, k]) for i in observation_range) >= 1)

@variable(model,
    d[i in observation_range, j in variable_range] >= 0)

@constraint(model,
    manhattan_distance_1[i in observation_range, j in variable_range, k in cluster_range],
    d[i, j] >= data[i, j] - c[k, j] - 1000*(1 - z[i, k]))
@constraint(model,
    manhattan_distance_2[i in observation_range, j in variable_range, k in cluster_range],
    d[i, j] >= c[k, j] - data[i, j] - 1000*(1 - z[i, k]))

@objective(model,
    Min,
    sum( sum( (d[i, j]) for j in variable_range) for i in observation_range))

optimize!(model)
```

```julia
optimal_z = Matrix(value.(z))
assignment = []
for i in 1:size(optimal_z)[1]
    for j in 1:size(optimal_z)[2]
        if (optimal_z[i, j] == 1)
            push!(assignment, j)
        end
    end
end
assignment

#value.(c) is DenseArray, shape: k x p
#value.(z) is DenseArray, shape: n x k
evaluations = evaluate_clustering(similarity_normalized, assignment, Matrix(value.(c)))
```
