```julia
using CSV, DataFrames, Gurobi, JuMP, LinearAlgebra, Plots, Clustering, StableRNGs, Random, Statistics, MultivariateStats
```

```julia
function read_data(file, scaling)
    # normalized_root = "./normalized_data"
    # scaled_root = "./scaled_data"
    
    if scaling == "normalized"
        data = CSV.read("./normalized_data/$file", DataFrame)
        for c in names(data, Any)
            data[!, c]= Float64.(data[!, c])
        end
        return data
    elseif scaling == "processed"
        data = CSV.read("./processed_data/$file", DataFrame)
        for c in names(data, Any)
            data[!, c]= Float64.(data[!, c])
        end
        return data
    else
        data = CSV.read("./scaled_data/$file", DataFrame)
        for c in names(data, Any)
            data[!, c]= Float64.(data[!, c])
        end
        return data
    end
end

function k_means_cluster(data, k, seed=0)
    # Documentation: https://clusteringjl.readthedocs.io/en/latest/kmeans.html
    
    data_matrix = Matrix(data)
    kmeans_result = Clustering.kmeans(data_matrix', k; rng=StableRNG(seed))
        # kmeans_result.totalcost (within-cluster sum of square distance)
        # a = assignments(kmeans_result) # get the assignments of points to clusters
        # c = counts(kmeans_result) # get the cluster sizes
        # M = kmeans_result.centers # get the cluster centers
    return kmeans_result
end
```

```julia
abalone_z = read_data("cc_general_mm.csv", "scaled");
abalone = read_data("cc_general.csv", "processed");
```

```julia
data_matrix = Matrix(abalone_z)

measure = []
K_max = 25
for k in 1:K_max
    Random.seed!(k)
    kmeans_result = Clustering.kmeans(data_matrix', k)
    push!(measure, sum(kmeans_result.totalcost))
end

plot(1:K_max, measure, xlabel="# of Clusters", ylabel="Within Cluster Sum of Squares", title="Elbow Plot for KMeans Clustering", linewidth=2)
```

```julia
##### Parameters #####
plotting = ""
data_z = abalone_z
data = abalone
dimension = "2D"
k_selected = 3
######################

abalone_kmeans = k_means_cluster(abalone_z, k_selected)
c_assignments = assignments(abalone_kmeans)
centroids = abalone_kmeans.centers'

if plotting == "scaled"
    PCA_model = fit(PCA, Matrix(data_z); maxoutdim=3)
    title = "Plotting PCs of Scaled Data"
else
    PCA_model = fit(PCA, Matrix(data); maxoutdim=3)
    title = "Plotting PCs of Standard Data"
end

println("PCA with dimensions $(size(PCA_model)[2])")
data_PCA = projection(PCA_model)

for k in 1:k_selected
    if dimension == "2D"
        if k == 1
            p = scatter(data_PCA[c_assignments .== k, 1], 
                        data_PCA[c_assignments .== k, 2], 
                        market=:circle, linewidth=0)
        else
            scatter!(data_PCA[c_assignments .== k, 1], 
                     data_PCA[c_assignments .== k, 2], 
                     market=:circle, linewidth=0)
        end
    else
        if k == 1
            p = scatter(data_PCA[c_assignments .== k, 1], 
                        data_PCA[c_assignments .== k, 2], 
                        data_PCA[c_assignments .== k, 3], 
                        market=:circle, linewidth=0)
        else
            scatter!(data_PCA[c_assignments .== k, 1], 
                     data_PCA[c_assignments .== k, 2], 
                     data_PCA[c_assignments .== k, 3], 
                     market=:circle, linewidth=0)
        end 
    end
end

if dimension == "2D"
    plot!(p, xlabel="PC1", ylabel="PC2", title=title)
else
    plot!(p, xlabel="PC1", ylabel="PC2", zlabel="PC3", title=title)
end
```
